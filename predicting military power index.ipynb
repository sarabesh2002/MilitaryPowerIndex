```python
def correlation(dataset,threshold):
    col_corr=set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>threshold:  
                colname=corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr    
```

```python
corr_features = correlation(x,0.85)
len(set(corr_features))
```

```python
import statsmodels.api as sm 
```

```python
names = list(x.columns.values)
```

```python
for i in range (0,len(names)):
    y1= x.loc[:,x.columns == names[i]]
    x1= x.loc[:,x.columns != names[i]]
    model = sm.OLS(y1,x1)
    results = model.fit()
    rsq = results.rsquared
    vif=round(1/(1-rsq),2)
    print("R square of {} column is {} keeping all the other column as features".format(names[i],round(rsq,2)))
    print("Variance inflation factor pf {} column is {}\n".format(names[i],vif))
    
```

# REGRESSION 

```python
x.head()
```

```python
y = wmd.iloc[:,0]
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=5)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
gradientregressor = GradientBoostingRegressor(max_depth=2,n_estimators=200,learning_rate=0.1)
```

```python
model=gradientregressor.fit(x_train,y_train)
```

```python
y_pred =model.predict(x_test)
r2_score(y_pred,y_test)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

feature_importance = model.feature_importances_

feature_importance = 100.0*(feature_importance/feature_importance.max())
sorted_idx=np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0])+0.5
plt.figure(figsize=(20,20))
plt.bar(pos,x.columns.values[sorted_idx])
plt.xlabel("relative importance")
plt.title("variable importance")
plt.show()
```
